```python
import pandas as pd
df = pd.read_csv('OnlineRetail.csv', encoding='ISO-8859-1', on_bad_lines='skip')
df.head()
```

```python
#Đọc bộ dữ liệu, cho biết số dòng, số cột và kiểu dữ liệu của các thuộc tính.
df.info()
```

```python
df.head()
```

```python
#Trích xuất dữ liệu các cột Description và Quantity lưu vào file OnlineRetail.csv
df_extracted = df[['Description', 'Quantity']]
df_extracted.to_csv('OnlineRetail.csv', index=False)
```

```python
#Trích xuất dữ liệu 1000 dòng đầu tiên lưu vào file OnlineRetail.xlsx
df_1000 = df.head(1000)
df_1000.to_excel('OnlineRetail.xlsx', index=False)
```

```python
#Trích xuất các bản ghi có số lượng từ 10 trở lên lưu vào file OnlineRetail.h5
df['Quantity'] = pd.to_numeric(df['Quantity'], errors='coerce')
df_filtered = df[df['Quantity'].fillna(0).astype(int) >= 10]
df_filtered.to_hdf('OnlineRetail.h5', key='df', mode='w')
```

```python
#Trích xuất dữ liệu các phần tử từ dòng 1000 đến dòng 2000, các cột Quantity, InvoiceDate, UnitPrice lưu vào file OnlineRetail.json
df_extracted = df.loc[1000:2000, ['Quantity', 'InvoiceDate', 'UnitPrice']]
df_extracted.to_json('OnlineRetail.json', orient='records', lines=True)
```
