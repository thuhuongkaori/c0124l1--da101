```python
#GDP của các quốc gia có đồng đều không
import pandas as pd
import matplotlib.pyplot as plt
plt.hist(data['GDP (millions of US$)'])
plt.title('Phân phối GDP của các quốc gia')
plt.xlabel('GDP (USD)')
plt.ylabel('Số lượng quốc gia')
plt.show()
#kết luận: không đồng đều
```

```python
#Mỗi châu lục có bao nhiêu quốc gia nằm trong bảng dữ liệu
continent_counts = data['Continent'].value_counts()
print(continent_counts)
```

```python
#Tổng GPD của các châu lục
continent_gdp = data.groupby('Continent')['GDP (millions of US$)'].sum()
print(continent_gdp)
```

```python
#Top 10 quốc gia có GDP cao nhất
top_10_gdp = data.sort_values(by='GDP (millions of US$)', ascending=False).head(10)
print(top_10_gdp)
```
