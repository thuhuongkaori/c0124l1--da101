```python
import pandas as pd
df=pd.read_csv('OnlineRetail.csv', encoding='latin')
df.head()
```

```python
#Đọc bộ dữ liệu, cho biết số dòng, số cột và kiểu dữ liệu của các thuộc tính.
df.info()
```

```python
#Xây dựng bảng Pivot table, với mỗi Số hóa đơn tính trung bình cộng số lượng các mặt hàng theo từng Quốc gia.
pivot_table=pd.pivot_table(df, values='Quantity', index='InvoiceNo', columns='Country', aggfunc='mean', fill_value=0)
print(pivot_table)
```

```python
#Xây dựng bảng Pivot table, với mỗi Khách hàng cho biết số lượng mua hàng lớn nhất và nhỏ nhất theo Kho.
pivot_table=pd.pivot_table(df, values='Quantity', index='CustomerID', columns='StockCode', aggfunc={min, max}, fill_value=0)
print(pivot_table)
```

```python
#Xây dựng bảng Pivot table, với mỗi Mã kho tính tổng số lượng các mặt hàng và trung bình cộng giá.
pivot_table_sum=pd.pivot_table(df, values='Quantity', index='StockCode', aggfunc='sum', fill_value=0)
pivot_table_mean=pd.pivot_table(df, values='UnitPrice', index='StockCode', aggfunc='mean', fill_value=0)
print(pivot_table_sum)
print(pivot_table_mean)
```

```python
#Xây dựng bảng Pivot table cho biết tổng số lượng hàng bán được của mỗi ngày.
df['InvoiceDate']=pd.to_datetime(df['InvoiceDate'])
pivot_table=pd.pivot_table(df, values='Quantity', index='InvoiceDate', aggfunc='sum', fill_value=0)
print(pivot_table)
```
